# Kernal Estimation
- 確率密度関数を推定する
- ほとんどヒストグラムを連続にしている感覚に近い

```python
import numpy as np
from scipy.stats import gaussian_kde
import matplotlib.pyplot as plt
from scipy.stats import norm
from sklearn.mixture import GaussianMixture
from scipy import integrate

np.random.seed(1)
```

```python
def true_func(x):
    return 0.5*norm.pdf(x, 2, 1)+0.5*norm.pdf(x, -2, 1)

# integrate.quad(mixtureGaussian, -np.inf, np.inf)
```

```python
x_difinition = np.arange(-5, 5, 0.01)

x_train = np.hstack([norm.rvs(2, 1, size=1000, random_state=0),norm.rvs(-2, 1, size=1000, random_state=0)])
kde = gaussian_kde(x_train)
pred = kde(x_difinition)
```

```python
def plot_result(x_difinition,x_train,pred):
    plt.title('kernel estimation')
    plt.scatter(x_train,[0]*len(x_train),color="black",s=5, marker="o",label= "train")
    plt.plot(x_difinition, pred,color="black",linestyle='dashed',label= "predict")
    plt.plot(x_difinition, true_func(x_difinition), color="black",linestyle='solid', label="true")
    plt.legend()
```

```python
plot_result(x_difinition, x_train, pred)
```
